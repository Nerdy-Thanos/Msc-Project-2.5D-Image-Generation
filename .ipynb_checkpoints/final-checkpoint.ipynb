```python
from matplotlib.pyplot import show
from torchvision.transforms import transforms
import torch
from torchvision.utils import make_grid
import os
from PIL import Image

from torchvision.io import read_image
import matplotlib.pyplot as plt

print(os.getcwd())
leftimg_path = "/Users/vignesh/Documents/GitHub/Msc-Project-2.5D-Image-Generation/driving/left"
import glob
cv_img = []
e = False
for img in glob.glob("/Users/vignesh/Downloads/archive/*.jpg"):
    n= Image.open(img)
    n = n.resize((1024,1024), Image.ANTIALIAS)
    transform = transforms.ToTensor()
    m = transform(n)
    cv_img.append(m)
    if len(cv_img) == 64:
        e = True
    if e is True:
        break
#plt.figure(figsize=(10,10))
plt.title("Holopix50k- Landscape Image Samples")
plt.imshow(make_grid(cv_img, nrow=2).permute(1, 2, 0), aspect="auto")
```
